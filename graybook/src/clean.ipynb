```python
import pandas as pd
```

```python
df = pd.read_csv("graybook-sept2021.csv")
df
```

```python
df["Present Salary"] = df["Present Salary"].replace('[\$,]', '', regex=True).astype(float)
df["Proposed Salary"] = df["Proposed Salary"].replace('[\$,]', '', regex=True).astype(float)
```

```python
df
```

```python
df.to_csv("graybook-sept2021-clean.csv", index=False)
```
