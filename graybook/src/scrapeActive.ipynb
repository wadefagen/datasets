```python
dept_code = None
dept_name = None
def f(row):
  global dept_code
  global dept_name

  if row["Job Title"] == row["Employee Name"]:
    dept_code, dept_name = row["Job Title"].split(" - ")
    row["Department Code"] = None
    row["Department Name"] = None
    return row
  else:
    row["Department Code"] = dept_code
    row["Department Name"] = dept_name
    return row
```

```python
import requests
collegeCodes = ["KL", "KY", "NQ", "LT", "LN", "NA", "KT", "NU", "KW", "KN", "MY", "NN", "KR", "KM", "KS", "KP", "LQ", "KU", "KV", "LB", "NP", "NS", "NB", "LF", "LP", "LG", "LL", "NT", "LR", "NL", "NJ", "NE", "LC", "NV"]
collegeNames = ["Agr, Consumer & Env Sciences", "Applied Health Sciences", "Auxiliary Units", "Carle Illinois Medicine", "Center Innov in Teach Learn", "Chancellor", "College of Media", "Div Intercollegiate Athletics", "Division of General Studies", "Education", "Energy Services", "Facilities", "Fine & Applied Arts", "Gies College of Business", "Graduate College", "Grainger Engineering", "Illinois International", "Law", "Liberal Arts & Sciences", "Medicine at UIUC", "OVCRI Institutes", "Ofc VC Inst Advancement", "Provost & VC Acad Affairs", "Public Safety", "School of Information Sciences", "School of Labor & Empl. Rel.", "School of Social Work", "Technology Services", "University Library", "VC Admin and Operations", "VC Student Affairs", "VC for Research and Innovation", "Veterinary Medicine", "Vice Chancellor for Diversity"]
```

```python

def processPage(rawHTML, collegeCode, collegeName):
  dfs = []
  h3s = rawHTML.split("<h3")
  for index in range(1, len(h3s)):
    h3_split = h3s[index]

    h3_full = "<h3" + h3_split
    h3_split = h3_split[ h3_split.find(">") + 1 : h3_split.find("</h3>") ]
    departmentName = h3_split.split("-")[1].strip()
    print(departmentName)

    df = pd.read_html(h3_full)[0]
    df["Department Name"] = departmentName
    df["College Code"] = collegeCode
    df["College Name"] = collegeName
    dfs.append(df)
  return pd.concat(dfs)
```

```python
from time import sleep

dfs = []
for i in range(len(collegeCodes)):
  print(f"=== {collegeName} ===")
  collegeCode = collegeCodes[i]
  collegeName = collegeNames[i]
  rawHTML = requests.get(f"https://www.trustees.uillinois.edu/trustees/resources/23-24-Graybook/{collegeCode}.html").text

  dfs.append( processPage(rawHTML, collegeCode, collegeName) )
  sleep(5)

df = pd.concat(dfs)
```

```python
df["Proposed Salary"] = df["Proposed Salary"].replace('[\$,]', '', regex=True).astype(float)
```

```python
df["Present Salary"] = df["Present Salary"].replace('[\$,]', '', regex=True).astype(float)
```

```python
df["Department Name"] = df["Department Name"].str.replace('&amp;', '&')
```

```python
df["College Name"] = df["College Name"].replace('&amp;', '&')
```

```python
def process(dfE):
  df_jobs = dfE[ dfE["Job Title"] != "Employee Total for All Jobs" ]
  
  jobTitles = list(df_jobs["Job Title"].unique())
  depts = list(df_jobs["Department Name"].unique())
  
  highest_salary_row = df_jobs.nlargest(1, "Proposed Salary")
  jobTitles.remove(highest_salary_row["Job Title"].values[0])
  depts.remove(highest_salary_row["Department Name"].values[0])

  if len(jobTitles) == 0:
    jobTitles = ""
  if len(depts) == 0:
    depts = ""

  return {
    "Employee Name": highest_salary_row["Employee Name"].values[0],

    "Primary Job Title": highest_salary_row["Job Title"].values[0],
    "Department Name": highest_salary_row["Department Name"].values[0],
    "College Code": highest_salary_row["College Code"].values[0],
    "College Name": highest_salary_row["College Name"].values[0],

    "Present FTE": dfE["Present FTE"].max(),
    "Proposed FTE": dfE["Proposed FTE"].max(),
    "Present Salary": dfE["Present Salary"].max(),
    "Proposed Salary": dfE["Proposed Salary"].max(),

    "Tenure": highest_salary_row["Tenure"].values[0],
    "Empl Class": highest_salary_row["Empl Class"].values[0],

    "Secondary Job Titles": str(jobTitles),
    "Secondary Department Names": str(depts),
  }


d = []
for name in df["Employee Name"].unique():
  dfSingle = df[ df["Employee Name"] == name ]

  dfAllJobs = dfSingle[ dfSingle["Job Title"] == "Employee Total for All Jobs" ]

  if len(dfAllJobs) > 1:
    values = dfAllJobs["Present Salary"].values

    val = values[0]
    identicalName = False
    for v in values:
      if v != val:
        identicalName = True

    if identicalName:
      if name == "Kumar, Praveen":
        indexFilter = [160, 161]
      elif name == "Li, Bo":
        indexFilter = [50, 51, 52, 232, 233]
      else:
        print("!!!")
        
      df1 = dfSingle[ dfSingle.index.isin(indexFilter) ]
      df2 = dfSingle[ ~dfSingle.index.isin(indexFilter) ]

      d.append(process(df1))
      d.append(process(df2))
    else:
      d.append(process(dfSingle))
  else:
    d.append(process(dfSingle))
```

```python
df_CLEAN = pd.DataFrame(d)
```

```python
from time import sleep


dfs = []

for url in urlList:
  x = pd.read_html(url)
  college_df = x[0]
  college = college_df.iloc[0, 0]
  print(college)
  college.split(" - ")
  college_code, college_name = college.split(" - ")

  college_df = college_df.drop(index=[0])
  college_df["College Code"] = college_code
  college_df["College Name"] = college_name

  college_df = college_df.apply(f, axis=1)
  college_df = college_df.dropna(subset=["Department Code"])
  dfs.append(college_df)
  sleep(5)
```

```python
df = pd.concat(dfs)
df.to_csv("graybook-2024.csv", index=False)
```
